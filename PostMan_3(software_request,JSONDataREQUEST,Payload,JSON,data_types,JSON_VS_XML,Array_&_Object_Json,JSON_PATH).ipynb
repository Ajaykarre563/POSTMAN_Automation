# **Payload**

# **Payload means the actual data you send in the body of an API request.**

**When you send data to the server (like username, password, etc.), that data is called the payload.**

It usually applies to POST, PUT, PATCH requests (not much for GET, because GET usually sends data in URL params).

Format is usually JSON, XML, or form-data.





```
You (Client) ----> (sending Request Payload) ----> Server (API)
Server (API) ----> (sending Response Payload) ----> You (Client)


```



**Request Payload = The data you send to the server.**

 Example: Sending student details to save.

**Response Payload = The data you get back from the server.**

 Example: Server confirms "Student created successfully".

# **Note:**

Files like .txt, .pdf, .docx, etc., are binary or document files. They can be large and unstructured, and they are not easy for a server to directly understand or process without special handling.

Servers usually prefer to work with structured, lightweight, and predictable data  that's why JSON is used.


#  **What is JSON?**

**Full form**: JavaScript Object Notation

**Purpose:** A syntax (format) for storing and exchanging data between systems (like client  server).

**Designed for:** Human-readable data exchange (easy for both humans and machines).

**Written in:** A text format based on JavaScript object style.

**Extended from**: JavaScript language, but now used everywhere (Python, Java, C#, etc.).

**Filename extension:** .json

**Internet media type (MIME type)**: application/json

**JSON stands for JavaScript Object Notation.**

**Its a lightweight and easy-to-read way to store and exchange data between a server and a client, or between two systems.**

**At its core, JSON is just text written in a structure that looks like a JavaScript object, with:**

Key-value pairs ("name": "John")

Arrays ("hobbies": ["reading", "cycling", "music"])

Nested objects ("address": {"city": "New York", "zip": "10001"})

# **Example of JSON:**



```
{
  "name": "Alice",
  "age": 25,
  "isStudent": false,
  "skills": ["Python", "SQL", "React"],
  "address": {
    "city": "London",
    "postcode": "W1A 1AA"
  }
}

```



# **Key points:**

Data is in key-value format.

Keys are always strings (inside double quotes "").

Values can be string, number, boolean, array, object, or null.

Very common in APIs and web development.

# **JSON Data Types:**

Number:
Example: 25, 3.14, -10

String (text inside double quotes ""):
Example: "Hello World", "Alice"

Boolean (true/false values):
Example: true, false

Null (empty or no value):
Example: null

# **What is a JSON Object?**
A JSON object is like a collection of key-value pairs.

It looks just like a Python dictionary or a JavaScript object.

The data is written inside { } curly braces.





```
{
  "name": "John",
  "age": 30,
  "isStudent": false
}

```



# **What is a JSON Array?**
A JSON array is an ordered list of values.

It looks just like a Python list.

The data is written inside [ ] square brackets.

Each item in an array can be a string, number, boolean, object, array, or null.

# **simple array example:**



```
[
  "apple",
  "banana",
  "cherry"
]

```



# **Array of objects example:**



```
[
  { "name": "John", "age": 30 },
  { "name": "Alice", "age": 25 },
  { "name": "Bob", "age": 27 }
]

```



# **What is JSONPath?**

JSONPath is like an address (or a query language) to find data inside a JSON file or object.

It is similar to XPath for XML, but for JSON.

It helps you easily access a specific part of a big JSON.
