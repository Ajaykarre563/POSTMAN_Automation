# **What is a Client?**(browsering the data through the internet)
A client is a computer hardware device or software that accesses a service made available by a server.

The server is often (but not always) on a different physical machine.

# **What is a Server?**(from where we get the data)
A server is a physical computer that runs services to fulfill requests from other computers (clients).

**Based on the service, it can be:**

File server

Database server

Home media server

Print server

Web server

# 🔹 **1-Tier Architecture:**
Everything is in one place — the client computer handles all tasks: UI, business logic, and data storage.

Example: A standalone desktop application like MS Excel or an offline database app.

# 🔹 **2-Tier Architecture:**
Split into:

Client Tier (Client computers)

Database Tier (Database server)

Clients interact directly with the database, usually for data retrieval and updates.

Example: A desktop app that connects to a remote SQL database.

# 🔹 **3-Tier Architecture:**
Divided into:

Client Tier – Users interact via a UI (e.g., browser or app).

Business Logic Tier – Handled by an application server that processes data and logic.

Database Tier – Stores and manages data.

More scalable, secure, and easier to maintain.

Example: Modern web applications (e.g., Amazon, Gmail).

# 📚 **Layered View of 3-Tier Architecture (Bottom Part of the Image):**

**Presentation Layer**

**Application Layer**

Tech: Java, Python, .NET, C#

Role: Contains business logic, processes user requests.

**Data Layer**

Tech: MySQL, PostgreSQL, Oracle, MongoDB

Role: Handles data storage and management.

# **what is an API (Application Program Interface)?**
An API allows two applications to interact and exchange data with each other, even if they are built using different platforms or technologies.

# 🔄 **How API Works (Using the Diagram):**
# **Three Layers:**

**Front-end of the application** (what the user interacts with)

**Back-end of the application** (where data is processed/stored)

**Middle-layer (API Layer)** – connects the front-end and back-end.

# **Responsibilities of the API Layer**:

📤 Fetch data from the back-end and show it on the front-end.

📥 Receive input from the front-end and send it to the back-end to store or process.

# **🔸 Types of APIs:**
# **SOAP (Simple Object Access Protocol)**

A protocol that uses XML-based messaging.

It’s more strict and standardized.

Often used in enterprise-level applications.

# **REST (Representational State Transfer)**

An architectural style, not a protocol.

Uses standard HTTP methods (GET, POST, PUT, DELETE).

More flexible and lightweight than SOAP.

# **✅ All Web Services are APIs:**
Web services use APIs to let two systems talk over the internet. So, a web service is a kind of API that works on a network.

# **❌ Not all APIs are Web Services:**
Some APIs don’t use the internet. For example, an API inside your computer software can help two programs talk without using a network—so, it’s not a web service.

# 💡 **In short:**
Web Service = API + Internet

API = Communication tool (with or without Internet)

# **Simple Definitions:**

# **URI (Uniform Resource Identifier)**

➤ A general name for something you can identify on the internet.
✅ It can be a URL or a URN.

# **URL (Uniform Resource Locator)**

➤ It's a type of URI that gives the exact address to access a resource (with scheme, host, path).
Example: https://google.com/articles/articlename

# **URN (Uniform Resource Name)**

➤ It's a name-only identifier; it doesn’t show how to locate the resource.
Example: urn:isbn:0451450523

🧠 Easy way to remember:

URI = Name OR Address

URL = Address

URN = Name

# **🔹 Feature vs Resource**
Feature → Used in manual testing (test some functionality manually).

Resource → Used in API automation testing (same thing, but tested using code).

# **🔹 Payload**
Payload = Body of the HTTP Request or Response.

Request Payload → Data sent by the client to the server (e.g., login info in JSON/XML).

Response Payload → Data sent back by the server to the client (e.g., success message, token, user data).

# **🔁 Client & Server Communication**
Client (S1) sends:

URI

Payload (body)

# **Server (S2) processes it and gives a:**

Status Code (e.g., 200 OK)

Response Payload (e.g., “Login Successful”)
