# **Workspace**

A workspace is an area where we maintain and save files.

You can create, rename, and delete a workspace.

# **Creating Collection**

A collection contains a number of folders and HTTP requests.

You can create, rename, delete, and run a collection.

You can create any number of collections under a workspace.

# **Method**:

GET → You are asking the server for some data (not sending any new data).

URL: https://reqres.in/api/users?page=2 → You are asking for users, specifically on page 2.

**Query Parameter:**

page = 2 → This is passed in the URL to tell the server which page of data you want.

**Response (below):**

Status Code: 200 OK → This means the request was successful.

Time: 1.78 seconds → The time taken by the server to respond.

Size: 2.18 KB → The size of the returned data.

**Response Body (the JSON part)**:

total: 12 → There are 12 users in total.

total_pages: 2 → All the users are spread across 2 pages.

data: (an array of user objects) → Each object has:


id: user's ID

email: user's email

first_name: user's first name

last_name: user's last name

avatar: link to the user's avatar image

# **response:**

**Body:**
Shows the actual data returned from the API (usually in JSON, XML, HTML, etc.).
In your earlier example, you saw the list of users here.

**Cookies:**
Shows any cookies that the server sends back.
Cookies are small pieces of data used to manage sessions, track users, etc.

**Headers (19):**
Lists the metadata information that comes with the response, like:

Content-Type

Server

Date

Content-Length (Here you have 19 headers in your response.)

**Test Results:**

If you write test scripts inside Postman (like checking if status code is 200), the results of those tests appear here.

Clock Icon (on the right):
Shows the response time and lets you check the request timeline.

# **What is a Status Code?**

A status code is a 3-digit number sent by the server when you make an API request (or visit a website).

It tells you whether the request succeeded or failed, and why.

# 🔵 **From your image:**

Level 200 → ✅ Success (everything worked)

Level 400 → ❌ Client Error (your mistake, like wrong URL, no token, etc.)

Level 500 → ⚠️ Server Error (server has a problem)
